```python
import pandas as pd
```

```python
#loading data sets into DF's
movies_df = pd.read_csv("movies_names.csv")
movies_tagline = pd.read_csv("movies_tagline.csv")
```

```python
# checking movies df
movies_df.head()
```

```python
#checking tagline df
movies_tagline.head()
```

```python
# left joining taglines into movies
movies_all = movies_df.merge(movies_tagline, on='id', how='left')
movies_all
```

```python
## getting toy story films from movie df into its own df
toy_story_df = movies_df[movies_df["original_title"].str.contains('Toy Story', regex=False)]
toy_story_df
```

```python
# merging toy story df with taglines
toy_story_tag = toy_story_df.merge(movies_tagline, on='id', how='left')
toy_story_tag
```

```python
# toy story merged df now have 5 columns including the tagline
toy_story_tag.shape
```

```python
movies_outer = movies_df.merge(movies_tagline, on='id', how='outer')
movies_outer
```
